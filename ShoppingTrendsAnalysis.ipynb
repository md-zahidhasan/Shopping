```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
### summary statistics using describe() method

df.describe(include= 'all')
```

```python
df.isnull().sum()
```

```python
### missing value replace of review ratings fillup by same category wise median value
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('median'))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
print(df.columns)
```

```python
### create a new column age_group 
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels= labels)
df[['age','age_group']].head()
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```
